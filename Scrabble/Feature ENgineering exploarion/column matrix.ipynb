### Transform to game level data frame

```python
def col_trans(data):
    # Create a new column 'Bot_nickname' based on the condition
    data['Bot_nickname'] = data['nickname'].apply(lambda x: x if 'Bot' in x else None)
    # Create a new column 'Player_nickname' based on the condition
    data['Player_nickname'] = data['nickname'].apply(lambda x: x if 'Bot' not in x else None)
    
    # Create new columns with default value of 0
    columns_to_transform = ['turns_count', 'max_point', 'min_point', 'avg_bingo', 'word_len', 'count_letters_JQXZ', 'rack_used_rate', 'win', 'game', 'rating', 'score']
    
    for col in columns_to_transform:
        # Update 'Bot_{col}' based on 'Bot_nickname'
        data[f'Bot_{col}'] = data.groupby(['game_id', 'Bot_nickname'])[col].transform(lambda x: x.iloc[0] if not x.isnull().all() else np.nan)
        
        # Update 'Player_{col}' based on 'Player_nickname'
        data[f'Player_{col}'] = data.groupby(['game_id', 'Player_nickname'])[col].transform(lambda x: x.iloc[0] if not x.isnull().all() else np.nan)
    
    # Drop unnecessary columns
    data = data.drop(['nickname'] + columns_to_transform, axis=1)
    
    # Aggregate data by game_id
    aggregated_data = data.groupby('game_id').agg({
        'Bot_nickname': 'first',
        'Player_nickname': 'first',
        'Bot_turns_count': 'sum',
        'Player_turns_count': 'sum',
        'Bot_max_point': 'sum',
        'Player_max_point': 'sum',
        'Bot_min_point': 'sum',
        'Player_min_point': 'sum',
        'Bot_avg_bingo': 'sum',
        'Player_avg_bingo': 'sum',
        'Bot_word_len': 'sum',
        'Player_word_len': 'sum',
        'Bot_count_letters_JQXZ': 'sum',
        'Player_count_letters_JQXZ': 'sum',
        'Bot_rack_used_rate': 'sum',
        'Player_rack_used_rate': 'sum',
        'Bot_win': 'sum',
        'Player_win': 'sum',
        'Bot_game': 'sum',
        'Player_game': 'sum',
        'Bot_score' : 'sum',
        'Player_score' : 'sum',
        'Bot_rating' : 'sum',
        'Player_rating' : 'sum',
    }).reset_index()

    return aggregated_data
```

```python
data_matix = col_trans(data_merge_all)
```
