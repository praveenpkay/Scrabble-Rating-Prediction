```python
import pandas as pd

def build_scrabble_board(game_df, game_id):
    # Filter the game data for the specified game ID
    game_data = game_df.loc[game_df['game_id'] == game_id]

    # Initialize the Scrabble board
    board = [['' for _ in range(15)] for _ in range(15)]

    # Place each move on the board
    for i, row in game_data.iterrows():
        move = row['move']
        rack = row['rack']

        # Ensure the turn_number is within a valid range
        if 1 <= row['turn_number'] <= 15:
            # Place the letters on the board
            for j, letter in enumerate(move):
                if letter != '.':
                    # Update the rack with the remaining letters
                    if j < len(board[row['turn_number'] - 1]):
                        board[row['turn_number'] - 1][j] = letter
                    else:
                        print(f"Warning: Turn {row['turn_number']} exceeds board length.")

    return board

# Example usage
game_id = 1
board = build_scrabble_board(games_turns_merged_train, game_id)
print(pd.DataFrame(board))
```

```python
import pandas as pd

def calculate_word_length_average(temp):
    total_words = 0
    total_word_length = 0

    for _, row in temp.iterrows():
        move = str(row['move'])  # Ensure the 'move' column is treated as a string

        # Split the move into words based on '.'
        words = move.split('.')
        
        # Calculate word length and update totals
        for word in words:
            if word:
                total_words += 1
                total_word_length += len(word)

    if total_words == 0:
        return 0

    return total_word_length / total_words

# Assuming 'game_df' is your DataFrame with the provided data
word_length_average = calculate_word_length_average(temp)
print(f'Word Length Average: {word_length_average:.2f}')
```

```python
def calculate_rack_usage_percentage(game_df):
    total_rack_letters = 0
    used_rack_letters = 0

    for _, row in game_df.iterrows():
        rack = str(row['rack'])  # Ensure the 'rack' column is treated as a string
        move = str(row['move'])  # Ensure the 'move' column is treated as a string

        # Count total rack letters
        total_rack_letters += len(rack)

        # Count used rack letters based on the move
        for letter in move:
            if letter != '.':
                used_rack_letters += 1

    if total_rack_letters == 0:
        return 0

    return (used_rack_letters / total_rack_letters) * 100

rank_usage = calculate_rack_usage_percentage(temp)
print(f'Rack Usage: {rank_usage:.2f}%')
```

```python
import pandas as pd

def calculate_high_scoring_letter_usage(temp):
    """Calculates the frequency of using high-scoring letters (Q, Z, J, X) in the DataFrame."""
    high_scoring_letter_count = 0
    total_letters = 0

    for _, row in temp.iterrows():
        move = str(row['move'])  # Ensure the 'move' column is treated as a string

        # Count high-scoring letters in the move
        high_scoring_letter_count += sum(1 for letter in move if letter in ['Q', 'Z', 'J', 'X'])
        
        # Count total letters in the move
        total_letters += len(move)

    if total_letters == 0:
        return 0

    return (high_scoring_letter_count / total_letters) * 100

# Assuming 'game_df' is your DataFrame with the provided data
high_scoring_letter_usage = calculate_high_scoring_letter_usage(temp)
print(f'High-Scoring Letter Usage: {high_scoring_letter_usage:.2f}%')
```
